```python
import requests
from bs4 import BeautifulSoup
import re

url = "http://unico2013.dothome.co.kr/crawling/exercise_bs.html"
r = requests.get(url)
r.encoding = "utf-8"
soup = BeautifulSoup(r.text, 'html.parser')
#print(soup)
h1_contents = soup.h1.string
atags = soup.find_all('a', limit=4)
img_src = soup.img['src']
h2_contents = soup.h2.string
ul_style_green = soup.find('li', style=re.compile('green$')).text
h2_2_contents = soup.find_all('h2')
oltags = soup.find_all('ol')
tabletags = soup.find_all('table')
tr_name = soup.find('tr', class_='name')
td_target = soup.find('td', id='target')

print(f"[<h1> 태그의 컨텐트] {h1_contents}")
print("[텍스트 형식으로 내용을 가지고 있는 <a> 태그의 컨텐트와 href 속성값]")
for tag in atags:
    print(f"{tag.text} : {tag['href']}")
print(f"[<img> 태그의 src 속성값] {img_src}")
print(f"[첫 번째 <h2> 태그의 컨텐트] {h2_contents}")
print(f"[<ul> 태그의 자식 태그들중 style 속성의 값이 green으로 끝나는 태그의 컨텐트] {ul_style_green}")
print(f"[두] 번째 <h2> 태그의 컨텐트] {h2_2_contents[1].text}")
print("[<ol> 태그의 모든 자식 태그들의 컨텐트]")
for tag in oltags:
    print(tag.text)
print("[<table> 태그의 모든 자손 태그들의 컨텐트]")
for tag in tabletags:
    print(tag.text)
print(f"[name이라는 클래스 속성을 갖는 <tr> 태그의 컨텐트] {tr_name.text}")
print(f"[target이라는 아이디 속성을 갖는 <td> 태그의 컨텐트] {td_target.text}")
```

```python
import requests
from bs4 import BeautifulSoup
import re

r = requests.get('http://unico2013.dothome.co.kr/crawling/exercise_bs.html')
r.encoding = 'utf-8'
html = r.text
bs = BeautifulSoup(html, 'html.parser')

print("[<h1> 태그의 콘텐츠]", bs.h1.string)
list = bs.find_all('a', text=True)
print(len(list))
print("[텍스트 형식으로 내용을 가지고 있는 <a> 태그의 콘텐츠와 href 속성값]")
for i in range(len(list)):
    print(list[i].string, ':', list[i].attrs['href'])
print("[<img> 태그의 src 속성값]", bs.img['src'])
print("[첫 번째 <h2> 태그의 콘텐츠]", bs.h2.string)
print("[<ul> 태그의 자식 태그들 중 style 속성의 값이 green으로 끝나는 태그의 콘텐츠]", bs.ul.find(style=re.compile('green$')).text)
print("[두 번째 <h2> 태그의 콘텐츠]", bs.find_all('h2')[1].string)
list2 = bs.ol.find_all()
print("[<ol> 태그의 모든 자식 태그들의 콘텐츠]")
for i in range(len(list2)):
    print(list2[i].string)
print("[<table> 태그의 모든 자손 태그들의 콘텐츠]", end="")
print(bs.table.text.strip())
print("[name이라는 클래스 속성을 갖는 <tr> 태그의 콘텐츠]", bs.find('tr', class_='name').text)
print("[target이라는 아이디 속성을 갖는 <td> 태그의 콘텐츠]", bs.find('td', id='target').text)
```
