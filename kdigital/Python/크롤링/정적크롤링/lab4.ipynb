```python
import requests
from bs4 import BeautifulSoup
import re

title = []
summary = []
rating = [] 

for n in range(1,11):
    req = requests.get('https://comic.naver.com/genre/bestChallenge?&page='+str(n))
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    titleList = soup.select('div.challengeInfo > h6 > a' )
    summaryList = soup.select('div.challengeInfo > div.summary')
    ratingList = soup.select('div.challengeInfo > div.rating_type > strong')
    for dom in titleList:
        content = dom.contents[0]
        content = re.sub("[\n\t]| ", '', content)
        title.append(content)
    for dom in summaryList:
        summary.append(dom.text)
    for dom in ratingList:
        rating.append(dom.text)

print(title)
print(summary)
print(rating)
```
