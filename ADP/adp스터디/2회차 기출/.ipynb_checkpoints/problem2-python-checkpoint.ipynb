#### 이상치 제거
- boxplot(IQR) 이용
- 결정값과 가장 상관성이 높은 피처들을 위주로 이상치를 검출하는 것이 좋다.

```python
# Co-relation 매트릭스
corr = solar[not_encoding_columns].corr()
# 마스크 셋업

# 그래프 셋업
plt.figure(figsize=(14, 8))
# 그래프 타이틀
plt.title('Overall Correlation of Titanic Features', fontsize=18)
#  Co-relation 매트릭스 런칭
sns.heatmap(corr, annot=False,cmap='RdYlGn', linewidths=0.2, annot_kws={'size':20})
plt.show()
```

- SOLAR THERMAL 변수를 통해 이상치 제거하기로 결정

```python
# 함수정의 => data와 index 정의하지만 불러오는 건 index
def outlier_iqr(data, column): 

    # lower, upper 글로벌 변수 선언하기    
    global lower, upper    
    
    # 4분위수 기준 지정하기     
    q25, q75 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)          
    
    # IQR 계산하기     
    iqr = q75 - q25    
    
    # outlier cutoff 계산하기     
    cut_off = iqr * 1.5          
    
    # lower와 upper bound 값 구하기     
    lower, upper = q25 - cut_off, q75 + cut_off     
    
    print('IQR은',iqr, '이다.')     
    print('lower bound 값은', lower, '이다.')     
    print('upper bound 값은', upper, '이다.')    
    
    # 1사 분위와 4사 분위에 속해있는 데이터 각각 저장하기     
    data1 = data[data[column] > upper]     
    data2 = data[data[column] < lower]
    data = pd.concat([data1,data2],axis=0)
    index = data.index
    
    # 이상치 총 개수 구하기
    print('총 이상치 개수는', data1.shape[0] + data2.shape[0], '이다.')
    return index
```

```python
outlier_index = outlier_iqr(solar,'SOLAR THERMAL')
```

```python
solar_df = solar.drop(outlier_index,axis=0)
solar_df
```

#### 데이터 분할

```python
feature_columns = list(solar_df.columns.difference(['SOLAR PV']))

X = solar_df[feature_columns]
y = solar_df['SOLAR PV']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

#### 스케일링

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

#### 예측 모델 생성
- 모델 성능 검증 : RMSE, R제곱, 정확도로 구하여라

- 정확도의 경우 실제값>예측값인 경우 (1-예측값/실제값), 실제값<예측값인 경우 (1- 실제값/예측값)으로 하고 이것들을 평균내라. 분수식의 분모가 0인 경우의 정확도는 0.5로 취급한다.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
clf = RandomForestRegressor()
clf.fit(X_train,y_train)

pred_test = clf.predict(X_test)
print('RMSE = ', np.sqrt(mean_squared_error(y_test,pred_test)))
print('R제곱 = ',r2_score(y_test,pred_test))
```

- RMSE 값은 522, R제곱은 0.94가 나왔다.

```python
def accuracy_score_personal(y,pred_y):
    global all_accuracy_score
    all_accuracy_score = []
    for i,j in zip(y,pred_y):
        if (i > j)&(i!=0):
            accuracy_score = 1-(j/i)
        elif (j > i)&(j!=0):
            accuracy_score = 1-(i/j)
        elif (j > i)&(j==0):
            accuracy_score = 0.5
        elif (i > j)&(i==0):
            accuracy_score = 0.5
    all_accuracy_score.append(accuracy_score)
    return np.mean(all_accuracy_score)
```

```python
accuracy_score_personal(y_test,pred_test)
```

- 정확도는 0.454이 나왔다.

#### 스케일링 된 데이터 활용
- 정확도가 상승하였다.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
clf = RandomForestRegressor()
clf.fit(X_train_scaled,y_train)

pred_test_scaled = clf.predict(X_test_scaled)
print('RMSE = ', np.sqrt(mean_squared_error(y_test,pred_test)))
print('R제곱 = ',r2_score(y_test,pred_test))
```

```python
accuracy_score_personal(y_test,pred_test_scaled)
```

#### 더 좋은 성능 위한 하이퍼파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

hyperparms = {
    'n_estimators':[100],
    'max_depth':[10,20],
    'max_features':[4,6,7],
    'min_samples_leaf':[8,12,18],
    'min_samples_split':[8,16,20],
}

gs = GridSearchCV(RandomForestRegressor(),
                 param_grid=hyperparms,
                 scoring='neg_mean_squared_error',
                 verbose=True,
                 n_jobs=-1)
```

```python
gs.fit(X_train,y_train)
```

```python
pred_test_gs = gs.predict(X_test)
print('RMSE = ', np.sqrt(mean_squared_error(y_test,pred_test_gs)))
print('R제곱 = ',r2_score(y_test,pred_test_gs))
print('정확도 = ',accuracy_score_personal(y_test,pred_test_gs))
```

- 정확도의 평가지수가 좋아졌다.
