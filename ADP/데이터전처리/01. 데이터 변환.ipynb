# 1장. 데이터 변환

## 데이터 set 불러오기

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
X = pd.DataFrame(iris.data,columns=iris.feature_names)
y = pd.DataFrame(iris.target,columns=["target"])
```

```python
X.shape, y.shape
```

```python
iris = pd.concat([X,y],axis=1)
iris.head()
```

## 01. np.where()
    
- 데이터 프레임 or 리스트 내 필드에 접근하고 수정하는 방법 . 
- np.where(조건)을 주면 해당 조건을 만족하는 인덱스를 반환한다.
- np.where(condition, x, y)는 조건이 충족되면x를, 그렇지 않으면y를 반환합니다.

```python
np.where(iris['sepal length (cm)']== 5.1)
```

```python
iris['Status'] = np.where(iris['sepal length (cm)']>=5, 'Good','Bad')
iris.head()
```

## 02. np.select()
- 데이터 프레임 or 리스트 내 필드에 접근하고 수정하는 방법.
- 조건목록과 선택목록을 리스트로 받아서 조건에 따라 선택 목록의 요소로 구성된 배열을 반환한다.
- 이 방법을 사용하면 2개 이상의 조건이 있을 때, Pandas의 주어진 조건을 기반으로 DataFrame을 만들 수 있다.
- 세 조건에 만족하지 않으면 defalt값으로 설정된다.(아래애서는 'Not Specified')

```python
conditionlist = [(iris['sepal length (cm)'] >= 5),
                 (iris['sepal length (cm)'] >= 4.5) & (iris['sepal length (cm)'] < 5),
                 (iris['sepal length (cm)'] < 4.5)]
choicelist = ['High', 'Mid', 'Low']
iris['status2'] = np.select(conditionlist, choicelist, default='Not Specified')
iris.head()
```

## 03-1. loc[ ]
- 명칭 기반으로 데이터 추출
- loc[행(인덱스)명칭, 열(컬럼)명칭] 으로 데이터 추출가능
- loc[조건,컬럼명] = data 로 데이터 변경 가능  => 조건을 &,| 등으로 여러개 줄 수 있음 **(데이터 전처리 시 많이 쓰이는 방법)**

```python
iris.loc[0,'sepal length (cm)']
```

```python
iris.loc[(iris['sepal length (cm)']==5.1), 'loc_data'] = "sepal length = 5.1"
iris.head()
```

```python
iris.loc[(iris['sepal length (cm)']==5.1)&(iris['sepal width (cm)']==3.5), 'loc_data2'] = "sepal length = 5.1 & sepal width = 3.5"
iris.head()
```

## 03-2. iloc[ ]
- 위치 기반 인덱싱으로 데이터 추출
- iloc[행(인덱스)위치,열(컬럼)위치] 으로 데이터 추출 가능
- iloc[행위치,열위치] = data 로 데이터변경 가능

```python
iris.iloc[0,1] = 3.51
iris.head()
```

```python
iris.iloc[0,1] = 3.5
iris.head()
```

#  04. replace()
- 문자열이나 숫자를 치환하는 함수이다. DataFrame에서도 활용이 가능
- inplace = True를 지정해주면 함수를 실행하면서 실제 값의 치환이 된다. (하지만 보통 df['a'] = df['a'].replace() 이런식으로 사용)
- **보통 특정 이상값이 있을 때 그 값을 모두 치환해줄 때 사용한다.(예 '-' 의 값을 np.NaN으로 모두 결측치 처리)**
    - .replace('-',np.NaN)

```python
# 데이터셋 새로 불러오기
iris.head()
```

```python
# 일반적인 값 변환
iris.replace('Good','Good!!',inplace=True)
iris.head()
```

```python
# 여러 값 변환
iris.replace({'Good!!':'Good','Bad':'Bad...'})
iris.replace(['Good!!','Bad'],['Good','Bad...'],inplace=True)
iris.head()
```

```python
# 여러 값을 한개의 값으로 치환
iris.replace(['Good','Bad...'],'Good or Bad')
```

```python
# 컬럼명 지정 문자열 변환
iris.replace({'Status': {'Good':'Good!!!'}}, inplace=True)
iris.head()
```

```python
# 컬럼명 지정 문자열 반환 - 어려 컬럼 값을 각각 문자열으로 변환
iris.replace({'Status':['Good!!!','Bad...'],'target':0},'예시!')
```

## 05. lambda

- 단일문으로 표현되는 익명함수 (한 줄로 표현되는 이름 필요없는 함수)
- 굳이 함수로 만들지 않고 1회성으로 만들어서 쓸 때 사용
- lambda 사용법 => lambda 입력인자 : 입력인자를 기반으로 한 계산식
- apply()가 가장 대표적으로 같이 쓰임
- filter(), map(), reduce() 등도 같이 유용하게 쓰일 수 있음

```python
# lambda 기본사용
lambda_squre = lambda x : x ** 2
print("3의 제곱은:", lambda_squre(3))
```

- apply() : 데이터프레임에 일괄적으로 함수를 적용시킬 때 쓰는 함수. 보통 아래와같이 lambda와 같이 쓰인다.

```python
# apply 예시1
iris['sepal length**2'] = iris['sepal length (cm)'].apply(lambda x:x**2)
iris.head()
```

```python
# apply 예시2 : 'bad'라는 값이 먼저 나오고 if이고, 'Good'은 else 뒤에 오는 것 주의!
iris['Status_lambda'] = iris['sepal length (cm)'].apply(lambda x:'bad' if x<5.0 else ('Good' if x>5 else 'normal'))
iris.head()
```

- filter, map, reduce와 사용

```python
# filter : 특정 조건을 만족하는 요소만 남기고 필터링

nums = [1, 2, 3, 6, 8, 9]
list(filter(lambda n:n%2==0, nums)) # True인 값들만 반환되는 로직이다.
```

```python
# map : 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트 반환

nums = [1, 2, 3, 6, 8, 9]
list(map(lambda n:n**2, nums))
```

```python
# reduce : 차례대로 앞 2개의 원소를 가지고 연산. 연산의 결과가 또 다음 연산의 입력으로 진행된다.
#          따라서 마지막까지 진행되면 최종 출력되는 한개의 값만 남게됨

import functools

a = [1, 3, 5, 8]
print(functools.reduce(lambda x,y:x+y, a)) # 1+3=4 -> 4+5=9 -> 9+8=17
print(functools.reduce(lambda x,y:x*y, a)) # 1*3=3 -> 3*5=15 -> 15*8=120
```
