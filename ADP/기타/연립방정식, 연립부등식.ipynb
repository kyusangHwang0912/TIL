```python
A = np.array([[2,-1], [1,-2]])
B = np.array([3,5])
C = np.linalg.solve(A,B)
print(C)
print(type(C))
```

# 부등식

1. 3x + y < 150
2. x + 2y < 100

```python
import numpy as np

x = np.arange(1, 1000)
y = np.arange(1, 1000)

results = []
for i in range(len(x)):
    for j in range(len(y)):
        if (3*x[i] + y[j] <= 150) and (x[i] + 2*y[j] <= 100):
            result = 2000*x[i]+3000*y[j]
            results.append([x[i],y[j],result])
        else:
            pass
```

```python
answer = []
for result in results:
    answer.append(result[2])
```

```python
max(answer)
```

```python
variables = []
for result in results:
    if result[2] == max(answer):
        print(result[0])
        print(result[1])
        print(result[2])
        variables.append((result[0],result[1]))
        
```
