# 시계열분석 부록 - 시각화함수

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from datetime import datetime
```

```python
df_apple = pd.read_csv('./data/apple_stock.csv', index_col='Date', parse_dates=True)
```

```python
# 바로 legend로 박힌다.
# 사실 Adj Close도 변화량을 가지고 있는데 변화량 없는것처럼 보인다.
df_apple[['Volume', 'Adj Close']].plot()
```

```python
# xlabel, ylabel, title 지정
df_apple[['Adj Close']].plot()
plt.ylabel('Close price')
plt.xlabel('overwite date')
plt.title('APPL')
```

```python
# volume의 y값을 오른쪽으로
df_apple[['Volume', 'Adj Close']].plot(secondary_y=['Volume'],figsize=(12,8))
```

```python
# 범위 지정해주기 1
df_apple['Adj Close']['2015-01-01':'2018-01-01'].plot()
```

```python
# 범위 지정해주기 2
df_apple['Adj Close'].plot(xlim=['2015-01-01','2018-01-01'],
                           ylim=[70,180])
```

```python
index = df_apple['2015-01-01':'2018-01-01'].index
stock = df_apple['2015-01-01':'2018-01-01']['Adj Close']
```

```python
f, ax = plt.subplots()
ax.plot_date(index,stock,'-')
f.autofmt_xdate() # 겹쳐서 나오는 것 알아서 해결해라
plt.tight_layout() # 플롯 겹칠 때 항상 써라
```

```python
# 다양한 시각화 major_locator, major_formatter 사용 - 논문쓸때 쓰이는 정도..
import matplotlib.dates as dates
f, ax = plt.subplots(figsize=(15,3))
ax.plot_date(index,stock,'-')
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b\n%Y'))
f.autofmt_xdate()
plt.tight_layout()
```
