```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from datetime import datetime
```

### 02-1. 데이터 불러오기

```python
# 바로 Date컬럼을 인덱스로 지정
df = pd.read_csv('./data/apple_stock.csv', index_col = 'Date')
```

```python
# datetime 형식으로 바꾸기
df.index = pd.to_datetime(df.index)
df.head()
```

---

### 02-2 shifting 적용

```python
# 뒤로 하나씩 shift한다. 대신 맨 앞 데이터가 missing이 되고, 마지막데이터는 업어짐
df.shift(1).head()
```

```python
# 마지막데이터는 삭제되고 두번째 데이터가 마지막데이터가 된 것
df.shift(1).tail()
```

```python
# 음수 shifting = 뒤의 데이터를 끌어오는 것

df.shift(-1).tail()
```

```python
# tshift 한달 뒤로 미루기??

df.tshift(freq='M', periods=1).head()
```
